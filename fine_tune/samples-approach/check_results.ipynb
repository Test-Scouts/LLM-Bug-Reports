```python
import pandas as pd
import pickle
```

```python
file_path_results = "data/results/mpnet-base/openoffice-results.pkl"
with open(file_path_results, 'rb') as f:
    mpnet = pickle.load(f)
```

```python
file_path_results = 'data/results/fine-tuned/openoffice-fine-tuned-results.pkl'
with open(file_path_results, 'rb') as f:
    fine_tuned = pickle.load(f)
```

```python
metric_values_mpnet = {
    "recall": {},
    "precision": {},
    "f1_score": {}
}
for key, value in mpnet.items():
    recall = value['true_positives'] / (value['true_positives']  + value['false_negatives'])
    precision = value['true_positives'] / (value['true_positives']  + value['false_positives'])
    f1_score = (2 * precision * recall) / (precision + recall)
    
    metric_values_mpnet["recall"][key] = recall
    metric_values_mpnet["precision"][key] =  precision
    metric_values_mpnet["f1_score"][key] =  f1_score


metric_values_fine_tuned = {
    "recall": {},
    "precision": {},
    "f1_score": {}
}
for key, value in fine_tuned.items():
    recall = value['true_positives'] / (value['true_positives']  + value['false_negatives'])
    precision = value['true_positives'] / (value['true_positives']  + value['false_positives'])
    f1_score = (2 * precision * recall) / (precision + recall)

    metric_values_fine_tuned["recall"][key] = recall
    metric_values_fine_tuned["precision"][key] = precision
    metric_values_fine_tuned["f1_score"][key] =  f1_score
```

```python
mpnet_df = pd.DataFrame(metric_values_mpnet)
# mpnet_df['reciprocal_recall'] = 1 / mpnet_df['recall']
mpnet_df.describe()
```

```python
fine_tuned_df = pd.DataFrame(metric_values_fine_tuned)
# fine_tuned_df['reciprocal_recall'] = 1 / fine_tuned_df['recall']
fine_tuned_df.describe()
```
